```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the CSV file
file_path = 'DATI_csv.csv'  # Replace with your actual file path
data = pd.read_csv(file_path, delimiter=';')

# Rename columns for clarity
data.rename(columns={'Unnamed: 0': 'Year'}, inplace=True)

# Extract the variables for regression
X = data['Revenue'].values.reshape(-1, 1)
y = data['Oncology_Spending'].values

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict values for the regression line
y_pred = model.predict(X)

# Plot the data points and the regression line
plt.figure(figsize=(10, 6), dpi=300)
plt.scatter(data['Revenue'], data['Oncology_Spending'], color='blue', label='Actual Data')
plt.plot(data['Revenue'], y_pred, color='red', label='Regression Line')
plt.title('Oncology Spending vs Revenue')
plt.xlabel('Revenue')
plt.ylabel('Oncology Spending')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr  # Importing pearsonr for correlation calculation

# Load the CSV file
file_path = 'DATI_csv.csv'  # Replace with your actual file path
data = pd.read_csv(file_path, delimiter=';')

# Rename columns for clarity
data.rename(columns={'Unnamed: 0': 'Year'}, inplace=True)

# Extract the variables for regression
X = data['Years'].values.reshape(-1, 1)  # Use 'Years' as the independent variable
y_revenue = data['Revenue'].values
y_oncology_spending = data['Oncology_Spending'].values

# Fit a linear regression model for both revenue and oncology spending
model_revenue = LinearRegression()
model_revenue.fit(X, y_revenue)

model_oncology = LinearRegression()
model_oncology.fit(X, y_oncology_spending)

# Predict values for both revenue and oncology spending
y_pred_revenue = model_revenue.predict(X)
y_pred_oncology = model_oncology.predict(X)

# Calculate Pearson's correlation coefficient
correlation_coefficient, _ = pearsonr(y_oncology_spending, y_revenue)

# Create a figure and axis objects
fig, ax1 = plt.subplots(figsize=(10, 6), dpi=300)

# Scatter plot for actual Oncology Spending (left y-axis)
ax1.scatter(data['Years'], data['Oncology_Spending'], color='blue', label='Actual World Oncology Spending')
ax1.plot(data['Years'], y_pred_oncology, color='blue', label='Regression Line for Oncology Spending')

# Set labels for the left y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('World Oncology Spending (bn)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for Revenue
ax2 = ax1.twinx()
ax2.scatter(data['Years'], data['Revenue'], color='red', label='Actual Revenue')
ax2.plot(data['Years'], y_pred_revenue, color='red', label='Regression Line for Revenue')

# Set labels for the right y-axis
ax2.set_ylabel('Revenue (bn)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add title
plt.title('Revenue and Oncology Spending from 2013 to 2023')

# Add grid lines for each year
ax1.grid(True, which='both', axis='x', linestyle='--', color='gray', alpha=0.5)  # Corrected grid line
ax1.set_xticks(data['Years'])  # Ensure grid lines are for each year
ax1.set_xticklabels(data['Years'], rotation=45)

# Add the Pearson correlation coefficient to the chart
ax1.text(0.05, 0.95, f"Pearson's Correlation Coefficient: {correlation_coefficient:.4f}",
         transform=ax1.transAxes, fontsize=12, verticalalignment='top', horizontalalignment='left',
         bbox=dict(facecolor='white', alpha=0.7, edgecolor='gray', boxstyle='round,pad=0.5'))

# Show the legend
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

# Show the plot
plt.show()

# Define the path to save the plot on your desktop
save_path = '/Users/TommasoMarrucci/Desktop/revenue_oncology_plot.png'

# Save the plot
fig.savefig(save_path)

# Confirm the path to the saved image
print(f"Plot saved at: {save_path}")
```
